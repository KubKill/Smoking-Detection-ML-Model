### 1. Import bibliotek

```python
!pip install detecto

import torch
import matplotlib.pyplot as plt
import numpy as np
import os

from detecto import core, utils, visualize
from detecto.visualize import show_labeled_image, plot_prediction_grid
from torchvision import transforms
```

```python
from google.colab import drive
drive.mount('/content/drive')
os.chdir('/content/drive/My Drive/Badanie')
```

### 2.Przygotowanie danych
#### ðŸ‡¬ðŸ‡§ Data preparation

```python
custom_transforms = transforms.Compose([
transforms.ToPILImage(),
transforms.Resize(900),
transforms.RandomHorizontalFlip(0.5),
transforms.ColorJitter(saturation=0.2),
transforms.ToTensor(),
utils.normalize_transform(),
])
```

### 3.Trenowanie modelu
#### ðŸ‡¬ðŸ‡§ Model training

```python
Train_dataset=core.Dataset('/content/drive/My Drive/Badanie/Train', transform=custom_transforms)
Test_dataset = core.Dataset('/content/drive/My Drive/Badanie/Test')
loader=core.DataLoader(Train_dataset, batch_size=1, shuffle=True)
model = core.Model(["palenie"])
losses = model.fit(loader, Test_dataset, epochs=30, lr_step_size=5, learning_rate=0.001, verbose=True)
```

```python
plt.plot(losses)
plt.show()
```

```python
image = utils.read_image("/content/drive/My Drive/Badanie/not_smoking_1.jpg") 
predictions = model.predict(image)
labels, boxes, scores = predictions
show_labeled_image(image, boxes, labels)
```

```python
thresh=0.6
filtered_indices=np.where(scores>thresh)
filtered_scores=scores[filtered_indices]
filtered_boxes=boxes[filtered_indices]
num_list = filtered_indices[0].tolist()
filtered_labels = [labels[i] for i in num_list]
show_labeled_image(image, filtered_boxes, filtered_labels)
```
